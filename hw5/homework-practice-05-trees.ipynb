# 1. Решающие деревья. Визуализация.

В этой части мы рассмотрим два простых двумерных датасета сделанных с помощью `make_moons`, `make_circles` и посмотрим как ведет себя разделяющая поверхность в зависимости от различных гиперпараметров.

```python
from sklearn.datasets import make_moons, make_circles, make_classification
datasets = [
    make_circles(noise=0.2, factor=0.5, random_state=42),
    make_moons(noise=0.2, random_state=42),
    make_classification(n_classes=3, n_clusters_per_class=1, n_features=2, class_sep=.8, random_state=3,
                        n_redundant=0, )
]
```

```python
palette = sns.color_palette(n_colors=3)
cmap = ListedColormap(palette)
```

```python
plt.figure(figsize=(15, 4))
for i, (x, y) in enumerate(datasets):
    plt.subplot(1, 3, i + 1)
    plt.scatter(x[:, 0], x[:, 1], c=y, cmap=cmap, alpha=.8)
```

```python
def plot_surface(clf, X, y):
    plot_step = 0.01
    palette = sns.color_palette(n_colors=len(np.unique(y)))
    cmap = ListedColormap(palette)
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                         np.arange(y_min, y_max, plot_step))
    plt.tight_layout(h_pad=0.5, w_pad=0.5, pad=2.5)

    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    cs = plt.contourf(xx, yy, Z, cmap=cmap, alpha=0.3)

    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap, alpha=.7,
                edgecolors=np.array(palette)[y], linewidths=2)
```

Постройте разделящие поверхности (для этого воспользуйтесь функцией `plot_surface`, пример ниже). Посчитайте accuracy на обучающей и тестовой выборках. Сильно ли деревья переобучились?

```python
# Пример:
from sklearn.linear_model import LinearRegression
X, y = datasets[2]
lr  = LinearRegression().fit(X, y)
plot_surface(lr, X, y)
```

```python
from sklearn.metrics import accuracy_score

for i, dataset in enumerate(datasets):
    X_train, X_test, y_train, y_test = train_test_split(*dataset)
    model = DecisionTreeClassifier().fit(X_train, y_train)
    plot_surface(model, X_test, y_test)
    plt.show()
    print("train accuracy = %.2f" % accuracy_score(y_train, model.predict(X_train)))
    print("test accuracy = %.2f" % accuracy_score(y_test, model.predict(X_test)))
```

__Задание 2. (1.5 балла)__

Попробуйте перебрать несколько параметров для регуляризации (напр. `max_depth`, `min_samples_leaf`). Для каждого набора гиперпараметров постройте разделяющую поверхность, выведите обучающую и тестовую ошибки. Можно делать кросс-валидацию или просто разбиение на трейн и тест, главное делайте каждый раз одинаковое разбиение, чтобы можно было корректно сравнивать (помните же, что итоговое дерево сильно зависит от небольшого изменения обучающей выборки?). Проследите как меняется разделяющая поверхность и обобщающая способность. Почему так происходит, одинаково ли изменение для разных датасетов?

```python
from sklearn.model_selection import GridSearchCV

params = {"max_depth": [1, 2, 4, 8, None],
          "min_samples_leaf": [1, 2, 3, 5]}
md_len = len(params["max_depth"])
msl_len = len(params["min_samples_leaf"])
for dataset in datasets:
    X_train, X_test, y_train, y_test = train_test_split(*dataset, train_size=0.8, shuffle=False)
    searcher = GridSearchCV(DecisionTreeClassifier(), params, scoring='accuracy').fit(X_train, y_train)
    plt.figure(figsize=(20, 20))
    for i, kit in enumerate(searcher.cv_results_["params"]):
        plt.subplot2grid((md_len, msl_len), (i // msl_len, i % msl_len))
        model = DecisionTreeClassifier(**kit).fit(X_train, y_train)
        plot_surface(model, *dataset)
        test_accuracy = accuracy_score(y_test, model.predict(X_test))
        train_accuracy = accuracy_score(y_train, model.predict(X_train))
        plt.title(f"{kit}" + "\n" + f"train accuracy = {train_accuracy}" + "\n" + f"test accuracy = {test_accuracy}")
    plt.show()
```

__Ответ:__
<br> Первый датасет: Данные представляют собой "круг" из объектов одного класса внутри "кольца" объектов другого класса. Маленькой глубины не хватает, чтобы построить хорошую разделяющую поверхность, а слишком глубокое дерево не улучшит результат или приведёт к переобучению.
<br> Второй датасет: Данные представляют собой два "вложенных" полумесяца, каждый из которых состоит из объектов одного класса. Как и в случае первого датасета, маленькой глубины дерева не хватает, а большая глубина избыточна.
<br> Третий датасет: Данные представляют собой три полосы, каждая из которых состоит из объектов одного класса; эти полосы пересекаются в одной точке. Значение глубины 2 является оптимальным, а при её увеличении дерево склонно к переобучению.
<br> Во всех случаях модели склонны к переобучению, если допустить листы размера 1, а оптимальное значение глубины деревьев зависит от геометрии данных.
